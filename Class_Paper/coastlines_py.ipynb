```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
fig, ax = plt.subplots(1,3,figsize=(30,15),subplot_kw={"projection":ccrs.PlateCarree()})

linewidth = 1.5

ax[0].coastlines(resolution='110m', linewidth=linewidth)
ax[1].coastlines(resolution='50m', linewidth=linewidth)
ax[2].coastlines(resolution='10m', linewidth=linewidth) 



# optionally zoom/pan:
for axis in ax:
    axis.set_extent((3, 15, 50, 71), crs=ccrs.PlateCarree())
    axis.add_feature(cfeature.LAND,color='cornflowerblue')
    
fig.suptitle('Coast of Norway', size = 40)

fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(5)  
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,12),subplot_kw={"projection":ccrs.PlateCarree()})

linewidth = 1.0

ax[0].coastlines(resolution='10m', linewidth=linewidth) 
ax[1].coastlines(resolution='10m', linewidth=linewidth) 
ax[0].set_extent((-9, 3, 45, 65), crs=ccrs.PlateCarree())
ax[1].set_extent((3, 15, 50, 70), crs=ccrs.PlateCarree())
# optionally zoom/pan:
for axis in ax:

    axis.add_feature(cfeature.LAND,color='cornflowerblue')
    
fig.suptitle('Coasts of Britain and Norway', size = 40)

fig.patch.set_edgecolor('black')
fig.patch.set_linewidth(5)  
```
